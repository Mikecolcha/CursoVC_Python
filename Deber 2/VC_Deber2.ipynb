OPERACIONES ARITMÉTICAS CON IMAGENES

MEZCLA

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img1 = cv.imread("Pic/nt.jpg")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2 = cv.imread("Pic/digih.jpg")
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)

# Igualar el tamano
fil,cols,chan = img1.shape
img2 = cv.resize(img2,(cols,fil))

# Mostrar imagenes
plt.figure(1)
plt.imshow(img1)
plt.title("Nanatsu")
plt.figure(2)
plt.imshow(img2)
plt.title("Digital")
```

```python
# Combinar imagenes
img_out = cv.addWeighted(img1,0.3,img2,0.2,0)
plt.imshow(img_out)
```

OPERACIONES BIT A BIT

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagenes
img1 = cv.imread('Pic/hat.jpg')
hat = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2 = cv.imread('Pic/digih.jpg')
dh = cv.cvtColor(img2, cv.COLOR_BGR2RGB)

# Mostrar las imagenes leidas
plt.figure(1)
plt.imshow(hat)
plt.figure(2)
plt.imshow(dh)
```

```python
# Cambiar el tamano de imagen
fil,col,_ = hat.shape
fil2,col2,_ = dh.shape
dh = cv.resize(dh,(col2//2,fil))
plt.imshow(dh)
```

```python
fil, col,_ = dh.shape
roi = img1[0:fil,450:450+col]
roi = cv.cvtColor(roi, cv.COLOR_BGR2RGB)
plt.imshow(roi)
```

```python
# Crear máscara del logotipo
dh_gray = cv.cvtColor(dh, cv.COLOR_RGB2GRAY)
ret,mask = cv.threshold(dh_gray, 150,255, cv.THRESH_BINARY)
plt.imshow(mask, cmap="gray")
```

```python
# Crear máscara invertida
mask_inv = cv.bitwise_not(mask)
plt.imshow(mask_inv, cmap="gray")
```

```python
# Tomar el roi menos la mascara
img1_bg = cv.bitwise_and(roi,roi,mask = mask)
plt.imshow(img1_bg)
```

```python
#tomamos region de intere del logotipo opencv
img2_bg = cv.bitwise_and(dh,dh,mask=mask_inv)
plt.imshow(img2_bg)
```

```python
img = img1_bg + img2_bg
plt.imshow(img)
```

```python
hat[0:fil,450:450+col]=img
plt.imshow(hat)
```
