CAPTURA DE EVENTOS DESDE EL RAT√ìN

```python
import cv2 as cv
clicked = False
# Funci√≥n encargada de capturar el click del mouse
def onMouse(evento,x,y,flags,param):
    # variable global
    global clicked
    if(evento ==cv.EVENT_LBUTTONUP):
        clicked =True
# Crear objeto cv.VideoCapture()
camCap = cv.VideoCapture(0)
cv.namedWindow("C√°mara")
# Configurar que se envie a la funcion onMouse los eventos del mouse
cv.setMouseCallback("C√°mara",onMouse)
print("Mostrando C√°mara, Presione en la Ventana para detener")
ret,frame = camCap.read()
while(ret and cv.waitKey(1)==-1 and not clicked):
    cv.imshow("C√°mara", frame)
    ret,frame = camCap.read()
camCap.release()
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
# Funci√≥n de dibujar c√≠rculos
def draw_circle_rectangule(event, x, y, flags, param):
    if(event == cv.EVENT_LBUTTONDOWN):
        cv.circle(img, (x, y), 25, (255,0,0), -1 )
# Funci√≥n de dibujar cuadrados
    elif(event == cv.EVENT_RBUTTONDOWN):
        cv.rectangle(img, (x, y), (x+50, y+50), (255,255,0), 2 )
# Crear
img = np.zeros((600,600,3), np.uint8)
cv.namedWindow('DIBUJANDO')
cv.setMouseCallback('DIBUJANDO', draw_circle_rectangule)
while(True):
    cv.imshow('DIBUJANDO',img)
    if (cv.waitKey(1) & 0xFF ==ord('q')):
        break
cv.destroyAllWindows()
```
